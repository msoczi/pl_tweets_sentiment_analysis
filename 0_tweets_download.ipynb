```python
# create api connection with twitter
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# load tweets ID's
tweets_ids = pd.read_csv('/content/drive/MyDrive/ZUM_NLP/pl_covid_tweets_clean.txt', sep='\t')
tweets_ids
```

```python
# Download tweets in buckets (50.000 each)
tweets = []
for k in range(11):
    for i in range(500*k,500*(k+1)):
        tweets_statuses = api.statuses_lookup(list(tweets_ids['tweet_id'][(i*100):(i*100+100)]))
        for i in range(len(tweets_statuses)):
            tweets.append(tweets_statuses[i].text)
    # Checklist 
    with open(f'/content/drive/MyDrive/ZUM_NLP/tweets_{50000*(k+1)}.pkl', 'wb') as f:
        pickle.dump(tweets, f)
    print(f'Pobrano {50000*(k+1)}/500000.')
    time.sleep(900) # sleep for 15 minuts - twetter limits
```
